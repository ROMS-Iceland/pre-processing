```python
# save initial encoding with fillvalue (only valid for BGC vraiables)
ini_encoding_north = ds['N3_n_north'].encoding
ini_encoding_east = ds['N3_n_east'].encoding
ini_encoding_west = ds['N3_n_west'].encoding
```

```python
ersem_varnames = list(ds.keys())
```

```python
new_varnames = [
    "B_OXYDEP_NUT", "B_OXYDEP_Oxy",
    "B_OXYDEP_DOM", "B_OXYDEP_POM",
    "B_OXYDEP_Phy", "B_OXYDEP_Het",
    "B_C_Alk", "B_C_DIC",
    "B_Ca_CaCO3", "B_CH4_CH4", "B_Bubble_Bubble",
    "B_BIOPLAST_MP_free", "B_BIOPLAST_MP_biof",
    "B_BIOPLAST_MP_het", "B_BIOPLAST_MP_det",
           ]
```

```python
def flatten(xss):
    return [x for xs in xss for x in xs]
```

```python
def name_gen(v):
    vnorth = v + "_north"
    veast = v + "_east"
    vwest = v + "_west"
    return [vnorth, veast, vwest]
```

```python
bry_names_ersem_1 = flatten([name_gen(v) for v in ["N3_n", "O2_o", "O3_TA", "O3_c"]])
bry_names_ersem_234 = name_gen("N3_n")
```

```python
bry_names_ersem_1
```

```python
bry_names1 = []
for n in ["B_OXYDEP_NUT", "B_OXYDEP_Oxy", "B_C_Alk", "B_C_DIC"]:
    bnames = name_gen(n)
    print(bnames)
    bry_names1.append(bnames)
```

```python
bry_names2 = []
for n in ["B_OXYDEP_DOM"]:
    bnames = name_gen(n)
    print(bnames)
    bry_names2.append(bnames)
```

```python
bry_names3 = []
for n in ["B_OXYDEP_POM", "B_OXYDEP_Phy", "B_OXYDEP_Het"]:
    bnames = name_gen(n)
    print(bnames)
    bry_names3.append(bnames)
```

```python
bry_names4 = []
for n in ["B_Ca_CaCO3", "B_CH4_CH4", "B_Bubble_Bubble"]:
    bnames = name_gen(n)
    print(bnames)
    bry_names4.append(bnames)
```

```python
bry_names5 = []
for n in ["B_BIOPLAST_MP_free", "B_BIOPLAST_MP_biof", "B_BIOPLAST_MP_het", "B_BIOPLAST_MP_det"]:
    bnames = name_gen(n)
    print(bnames)
    bry_names5.append(bnames)
```

```python
dict1 = dict(zip(flatten(bry_names1), [ds[v] for v in bry_names_ersem_1]))
dict2 = dict(zip(flatten(bry_names2), [ds[v]*0 + 1.0 for v in bry_names_ersem_234]))
dict3 = dict(zip(flatten(bry_names3), [ds[v]*0 + 0.001 for v in bry_names_ersem_234*3]))
dict4 = dict(zip(flatten(bry_names4), [ds[v]*0.0 for v in bry_names_ersem_234*3]))
dict5 = dict(zip(flatten(bry_names5), [ds[v]*0.0 for v in bry_names_ersem_234*4]))
```

```python
dict(zip(flatten(bry_names5), bry_names_ersem_234*3))
```

```python
dict1
```

```python
# assign new values (all attrs and encoding will be deleted)
ds = ds.assign(dict1)
ds = ds.assign(dict2)
ds = ds.assign(dict3)
ds = ds.assign(dict4)
ds = ds.assign(dict5)
```

```python
ds = ds.drop_vars(ersem_varnames)
```
